```python
%matplotlib inline
import sympy as sm
from sympy.physics.continuum_mechanics.beam import Beam
```

```python
E = sm.Symbol('E')
I = sm.Symbol('I')
b = Beam(10, E, I)
r0, m0 = b.apply_support(0, type="fixed")
r10 = b.apply_support(10, type="pin")
b.solve_for_ild_reactions(-1, r0, m0, r10)
```

```python
b.plot_ild_reactions();
```
