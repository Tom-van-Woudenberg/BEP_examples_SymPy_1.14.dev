```python
%matplotlib inline
import sympy as sm
from sympy.physics.continuum_mechanics.beam import Beam
```

```python
E = sm.Symbol('E')
I = sm.Symbol('I')
b = Beam(8, E, I)
r0, m0 = b.apply_support(0, type='fixed')
r8 = b.apply_support(8, type='pin')
b.apply_rotation_hinge(5)
b.apply_load(-10, 5, -1)
b.solve_for_reaction_loads(r0, m0, r8)
```

```python
b.plot_bending_moment();
```
